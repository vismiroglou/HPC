```python
import numpy as np
```

```python
pre = 5000
pim = 5000
re_floor = -2
re_ceiling = 1
im_floor = -1.5
im_ceiling = 1.5
T = 2
I = 100
```

```python
re = np.linspace(re_floor, re_ceiling, pre)
im = np.linspace(im_floor, im_ceiling, pim)
X, Y = np.meshgrid(re, im)
c = X + Y * 1j
z = np.zeros_like(c)
img = np.zeros((pre, pim))
```

```python
for i in range(I):
    mask = np.abs(z) < 2
    z[mask] = z[mask] * z[mask] + c[mask]
    img += mask.astype(int)
```

```python
from matplotlib import pyplot as plt
figure = plt.figure(figsize=(5,5), dpi=500)
plt.imshow(img, cmap='hot', extent=[re_floor, re_ceiling, im_floor, im_ceiling])
plt.xlabel("Re[c]")
plt.ylabel("Im[c]")
plt.title("M(c)")
plt.show()
```
